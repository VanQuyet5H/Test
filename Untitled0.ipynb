```python
#Hệ chữ kí el-Gamal
p=int(input('Nhập vào giá trị p:'))
a=int(input('Nhập vào giá trị a:'))
x=int(input('Nhập vào giá trị x:'))
k=int(input('Nhập vào giá trị k:'))
def ext_gcd(p,k):
	A3,B3=p,k
	A1=1
	B1=0
	A2=0
	B2=1
	while(B3!=0):
		Q=A3//B3
		R3=A3%B3
		Ar=A1-Q*A2
		Br=B1-Q*B2
		A3=B3
		A1=A2
		B1=B2
		B3=R3
		A2=Ar
		B2=Br
	return B1
T=ext_gcd(p-1,k)
y=(a**x)%p
print("Khóa công khai là:",p,a,y);
print("Khóa bí mật là:",x)
X=int(input('Nhập vào X:'))
r=(a**k)%X
s=T*(X-x*r)%(p-1)
print("TIN với chữ ký:",r,s)
if (y**r)*(r**s)%p == (a**X) % p:
  ver=True
else: 
  if (y**r)*(r**s)%p != (a**X) % p:
    ver=False
print("Xác minh:",ver)
```

```python
alphabet='abcdefghijklmnopqrstuvwxyz'
letter_to_index=dict(zip(alphabet,range(len(alphabet))))
index_to_letter=dict(zip(range(len(alphabet)),alphabet))
def encrypt(message,key):
	encrypted=''
	split_message=[message[i:i+len(key)] for i in range(0,len(message),len(key))]
	for each_split in split_message:
		i=0
		for letter in each_split:
			number=(letter_to_index[letter]+letter_to_index[key[i]]) % len(alphabet)
			encrypted+= index_to_letter[number]
			i+=1
	return encrypted
def decrypt(cipher,key):
	decrypted=''
	split_cipher=[cipher[i:i+len(key)] for i in range(0,len(cipher),len(key))]
	for each_split in split_cipher:
		i=0
		for letter in each_split:
			number=(letter_to_index[letter]-letter_to_index[key[i]])%len(alphabet)
			decrypted+=index_to_letter[number]
			i+=1
	return decrypted

def main():
	key='bimat'
	message='dungmotkehoach'
	encrypted_message=encrypt(message,key)
	decryted_message=decrypt(encrypted_message,key)
	print('Original message:'+ message)
	print('encrypted message:'+ encrypted_message)
	print('decryted_message:'+decryted_message)
main()
```
